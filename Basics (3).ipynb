```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(12, 25))

for sp in range(0,18,3):
    index = int(sp/3)
    ax1 = fig.add_subplot(6,3,sp+1)
    ax1.plot(women_degrees['Year'], women_degrees[stem_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax1.plot(women_degrees['Year'], 100-women_degrees[stem_cats[index]], c=cb_orange, label='Men', linewidth=3)
    for key,spines in ax1.spines.items():
        spines.set_visible(False)
    ax1.set_xlim(1968, 2011)
    ax1.set_ylim(0,100)
    ax1.set_title(stem_cats[index])
    ax1.tick_params(bottom="off", top="off", left="off", right="off", labelbottom="off")
    ax1.set_yticks([0,100])
    ax1.axhline(50, c=(89/255,89/255,89/255), alpha=0.5)
    
    if index == 0:
        ax1.text(2005, 87, 'Men')
        ax1.text(2002, 8, 'Women')
    elif index == 5:
        ax1.text(2005, 62, 'Men')
        ax1.text(2001, 35, 'Women')
        ax1.tick_params(labelbottom="on")






for sp in range(1,16,3):
    index = int(sp/3)
    ax2 = fig.add_subplot(6,3,sp+1)
    ax2.plot(women_degrees['Year'], women_degrees[lib_arts_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax2.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[index]], c=cb_orange, label='Men', linewidth=3)
    for key,spines in ax2.spines.items():
        spines.set_visible(False)    
    ax2.set_xlim(1968, 2011)
    ax2.set_ylim(0,100)
    ax2.set_title(lib_arts_cats[index])
    ax2.tick_params(bottom="off", top="off", left="off", right="off", labelbottom="off")
    ax2.set_yticks([0,100])
    ax2.axhline(50, c=(89/255,89/255,89/255), alpha=0.5)
    
    if index == 0:
        ax2.text(2005, 87,'Men')
        ax2.text(2002, 8,'Women')
    elif index == 4:
        ax2.tick_params(labelbottom="on")



for sp in range(2,20,3):
    index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[index]], c=cb_orange, label='Men', linewidth=3)
    for key,spines in ax.spines.items():
        spines.set_visible(False)    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom="off")
    ax.set_yticks([0,100])
    ax.axhline(50, c=(89/255,89/255,89/255), alpha=0.5)
    
    if index == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom="on")

    
plt.show()

   
```
